```python
my_age = int(input("please enter your age"))
my_score = float(input("please enter your exam score"))
print("So, your age is", my_age, "years old and you have scored", my_score, "Marks")
print("So, your age is %d years old and you have scored %0.2d Marks" %(my_age, my_score))
print(f"So, your age is {my_age} years old and have scored {my_score:0.2f} Marks")
```
